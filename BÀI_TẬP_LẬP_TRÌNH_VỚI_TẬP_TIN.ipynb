```python
import random
a = []
for i in range (0,1000):
    a.append(str(random.randint(-1000,1000)))
e = input("nhap ten tap tin: ")
d = open(e, "w")
c = []
for i in range(0,10):
    c.append(i)
for j in range(0,100):
    for k in range(0,10):
        c[k]=a[j*10+k]
    d.write(",".join(c)+ "\n")
d.close()
d = open(e, "r")
p = d.readlines()
for i in range(0,100):
    p[i]="  ".join(p[i].split(","))
    print(p[i])
```
